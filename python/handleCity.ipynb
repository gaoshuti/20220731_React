```python
# 获得每月的天数
def getDays(year,month):
#     year=int(dateStr[:4])
#     month=int(dateStr[4:])
    if month in [1,3,5,7,8,10,12]:
        return 31
    elif month in [4,6,9,11]:
        return 30
    else:
        if year%4!=0:
            return 28
        elif year%100==0 and year%400!=0:
            return 28
        else:
            return 29
def normalDate(year,month,day):
    if month < 10:
        month = '0'+str(month)
    if day < 10:
        day = '0'+str(day)
    return str(year)+'-'+str(month)+'-'+str(day)
i = 0
i2day = {}
day2i = {}
for year in range(2011,2019):
    for month in range(1,13):
        for day in range(1,getDays(year,month)+1):
            date = normalDate(year,month,day)
            i2day[i] = date
            day2i[date] = i
            i += 1
print(len(i2day))
print(i2day)
```
