```python
import json
from pyexcel_xlsx import get_data
from os import listdir
import pandas as pd


sourcePath = '/Users/rumeng/Downloads/上市公司基本信息年度表/STK_LISTEDCOINFOANL.xlsx'
'''
1.* Stkcd [证券代码] - 以上交所、深交所公布的证券代码为准
2.* Trddt [交易日期] - 以YYYY-MM-DD表示
3.* Opnprc [日开盘价] - A股以人民币元计，上海B以美元计，深圳B以港币计
4.* Hiprc [日最高价] - A股以人民币元计，上海B以美元计，深圳B以港币计
5.* Loprc [日最低价] - A股以人民币元计，上海B以美元计，深圳B以港币计
6.* Clsprc [日收盘价] - A股以人民币元计，上海B以美元计，深圳B以港币计
7.* Dnshrtrd [日个股交易股数] - 0=没有交易量
8.* Dnvaltrd [日个股交易金额] - 计量货币：人民币元。A股以人民币元计，上海B以美元计，深圳B以港币计，0=没有交易量
9.* Dsmvosd [日个股流通市值] - 个股的流通股数与收盘价的乘积，A股以人民币元计，上海B股以美元计，深圳B股以港币计，注意单位是千
10.* Dsmvtll [日个股总市值] - 个股的发行总股数与收盘价的乘积，A股以人民币元计，上海B股以美元计，深圳B股以港币计，注意单位是千
11.* Dretwd [考虑现金红利再投资的日个股回报率] - 上市首日的前收盘价取招股价,字段说明见“回报率计算方法”
12.* ChangeRatio [涨跌幅] - 基于昨收盘(交易所)计算的涨跌幅，日收盘价/昨收盘(交易所)-1

13.MarkettypeID [股票市场类型编码] - P9714：综合A股和创业板和科创板
14.RiskPremium1 [市场风险溢价因子(流通市值加权)] - 日度，考虑现金红利再投资的日市场回报率(流通市值加权平均法)与日度化无风险利率之差（央行公布三月定存基准利率）。
15.SMB1 [市值因子(流通市值加权)] - 日度，小盘股组合和大盘股组合的收益率之差，组合划分基于FAMA 2*3组合划分方法。组合日收益率的计算采用流通市值加权计算。
16.HML1 [账面市值比因子(流通市值加权)] - 日度，高账面市值比组合和低账面市值比组合的收益率之差，组合划分基于FAMA 2*3组合划分方法。组合投资收益率的计算采用流通市值加权。

17.UMD1 [动量因子（流通市值加权）] - 月度，Carhart动量因子计算方法：形成期为11个月，累积收益最高的前30%的投资组合收益率-累积收益最低的30%的投资组合收益率（流通市值加权）

18.* Proportion [机构持股比例] - 季度，机构投资者持有股份占股票流通股本数量的比例

19.portReturn [组合收益率] - 日度
20.turnover [组合换手率] - 日度
21.liquidity [流动性] - 日度

[流通市值] - 日度，每个城市所有股票的流通市值平均值
[市值账面价值比] - 
'''
# data = get_data(sourcePath)
# # print(type(data))     
# # 可以看到, 类型为collections.OrderedDict
# # print(json.dumps(data))
# # print(data['sheet1'])
# abbrTitle=data['sheet1'][0]
# titles=data['sheet1'][1]
# info=data['sheet1'][3:]
# print(abbrTitle)
# print(titles)
# print(info[0])
```

```python
citys=['深圳','上海','广州','北京','成都','杭州','重庆','西安','苏州','武汉','南京','天津','郑州','长沙','东莞','佛山','宁波','青岛','沈阳','合肥']
def handleAddress(address):
    for i in citys:
        if i in address:
            return i
    if '市' in address:
        address = address[:address.find('市')+1]
    elif '区' in address:
        address = address[:address.find('区')+1]
    elif '州' in address:
        address = address[:address.find('州')+1]
    elif '自治' in address:
        address = address[:address.find('自治')+3]
    elif address=='':
        print('为空')
    else:
        print(address)
    return address
```

```python
## 处理属性19-21 portReturn [组合收益率] turnover [组合换手率]  liquidity [流动性] - 日度
## 处理属性 组合流通市值 MarCap
sourcePath7='/Users/rumeng/Downloads/qingzang/trainSet/stock/'
objPath7='/Users/rumeng/Downloads/qingzang/trainSet/explainedVariable/'
cities=['深圳','上海','广州','北京','成都','杭州','重庆','西安','苏州','武汉','南京','天津','郑州','长沙','东莞','佛山','宁波','青岛','沈阳','合肥']
# cities=['深圳']
riskFreeInterestRate = {}
portReturnDict = {}
dates = []
filelist = listdir(sourcePath7)

sampleStkList = []

numofstocks=0

df = pd.read_excel('/Users/rumeng/Downloads/无风险利率/TRD_Nrrate.xlsx')
data=df.values
for i in range(2,len(data)):
    riskFreeInterestRate[data[i][1]]=data[i][3]
# print(riskFreeInterestRate)
for city in cities:
    for stkcd in stockId2CityDict[city]:
        if stkcd+'.json' not in filelist:
            continue
        sampleStkList.append(stkcd)
        with open(sourcePath7+stkcd+'.json', 'r',encoding='utf-8-sig') as f:
            lines = f.readlines()
            numofstocks+=1
            for line in lines:
                stockInfo=json.loads(line)
                date=stockInfo['Trddt']
                rfir=riskFreeInterestRate[date]
                if date not in portReturnDict.keys():
                    dates.append(date)
                    portReturnDict[date]={'Trddt':date,
                                          'RFIR':rfir,
                                          'sum1':(stockInfo['Dretwd']-rfir)*stockInfo['Dsmvosd']*1000,
                                          'sum2':stockInfo['Dnshrtrd']*stockInfo['Clsprc'],
                                          'sum3':abs(stockInfo['Dretwd']*stockInfo['Dsmvosd']*1000/stockInfo['Dnshrtrd']),
                                          'Dsmvosd':stockInfo['Dsmvosd']*1000,
                                          'sum4':stockInfo['Dsmvosd']*stockInfo['Dsmvosd']*1000,
                                          'portReturn':'',
                                          'turnover':'',
                                          'liquidity':'',
                                          'MarCap':''}
                else:
                    portReturnDict[date]['sum1']+=(stockInfo['Dretwd']-rfir)*stockInfo['Dsmvosd']*1000
                    portReturnDict[date]['sum2']+=stockInfo['Dnshrtrd']*stockInfo['Clsprc']
                    portReturnDict[date]['sum3']+=abs(stockInfo['Dretwd']*stockInfo['Dsmvosd']/stockInfo['Dnshrtrd']*1000)
                    portReturnDict[date]['Dsmvosd']+=stockInfo['Dsmvosd']*1000
                    portReturnDict[date]['sum4']+=stockInfo['Dsmvosd']*stockInfo['Dsmvosd']*1000
                    
    with open(objPath7+city+'.json', 'a',encoding='utf-8-sig') as json_file:    
        for date in dates:
            portReturnDict[date]['portReturn']=portReturnDict[date]['sum1']/portReturnDict[date]['Dsmvosd']
            portReturnDict[date]['turnover']=portReturnDict[date]['sum2']/portReturnDict[date]['Dsmvosd']
            portReturnDict[date]['liquidity']=portReturnDict[date]['sum3']/portReturnDict[date]['Dsmvosd']
            portReturnDict[date]['MarCap']=portReturnDict[date]['sum4']/portReturnDict[date]['Dsmvosd']
            j = json.dumps(portReturnDict[date],ensure_ascii=False)  
            json_file.write(j)
            json_file.write('\n') 
```
